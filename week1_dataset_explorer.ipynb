```python
clubs = df['current_club_name'].unique().tolist()
```

```python
club_player_count = {
    club: len(df[df['current_club_name'] == club])
    for club in clubs[:5]
}
```

```python
for club, count in club_player_count.items():
    print(club, count)
```

```python
def value_category(value):
    if value > 50000000:
        return "High Value"
    elif value > 10000000:
        return "Medium Value"
    else:
        return "Low Value"

df['value_category'] = df['market_value_in_eur'].apply(value_category)
```
