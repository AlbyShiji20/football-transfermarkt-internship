```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./data/players_cleaned.csv")
```

```python
df.describe()
df['market_value_in_eur'].describe()
df['last_season'].describe()
```

```python
plt.hist(df['market_value_in_eur'], bins=50)
plt.title('Distribution of Market Value in EUR')
plt.xlabel('Market Value in EUR')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x=df['market_value_in_eur'])
plt.title('Market Value Outliers')
plt.show()
```

```python
df[['last_season', 'market_value_in_eur']].corr()
```

```python
sns.heatmap(df[['last_season', 'market_value_in_eur']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
trend = df.groupby('last_season')['market_value_in_eur'].mean()

plt.plot(trend.index, trend.values)
plt.xticks(rotation=45)
plt.title('Average Market Value Over Seasons')
plt.xlabel('Season')
plt.ylabel('Average Market Value in EUR')
plt.show()
```

```python
top_clubs = df.groupby('club')['market_value_in_eur'].mean().sort_values(ascending=False).head(10)

top_clubs.plot(kind='bar')
plt.title("Top 10 Clubs by avg market value ")
plt.ylabel("Avg Market Value")
plt.show()
```

```python
sns.kdeplot(df['market_value_in_eur'], shade=True)
plt.title("KDE of Market Value")
plt.show()
```

### Key Insights
1. Market values have steadily increased over recent seasons.
2. A few elite clubs dominate the high-value player market.
3. Market values are right-skewed, indicating few very expensive players.
