```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./data/players.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['market_value_in_eur_'] = df['market_value_in_eur'].fillna(df['market_value_in_eur'].median())
```

```python
df['market_value_in_eur_'] = df['market_value_in_eur'].astype(float)
df['last_season'] = df['last_season'].astype(int)
```

```python
df.rename(columns={
    'name': 'player',
    'current_club_name': 'club'
}, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.to_csv("./data/players_cleaned.csv", index=False)
```

```python
df[df['last_season'] < 2020]
```

```python
df.sort_values(by='market_value_in_eur_', ascending=False).head(10)
```

```python

df.groupby('current_club_name')['market_value_in_eur_'].mean().idxmax()
```

```python
df[['last_name', 'market_value_in_eur_']].sort_values(by='market_value_in_eur_', ascending=False).head(10)
```

```python
df.groupby('current_club_name')['market_value_in_eur_'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('country_of_birth')['market_value_in_eur_'].mean().sort_values(ascending=False).head(10)
```

```python
df.groupby('last_season')['market_value_in_eur_'].mean()
```

### Key Insights
- Real Madrid Club de Fútbol, Manchester City Football Club, Paris Saint-Germain Football Club,  Futbol Club Barcelona, Arsenal Football Club have higher average market value
- Certain clubs consistently dominate player valuation
- Market values peaked in specific seasons due to transfer inflation
